```python
# Concatenating the two Datasets
#frames = [datapop_2016, datapop_2022]
#data_pop = pd.concat(frames, ignore_index=True)
#data_pop.head()
```

```python
# Perform the merge 'inner'
newdata_df = pd.merge(datapop_2016, datapop_2022, how ='inner', on=['County', 'Marital Status', 'Sex', 'UNIT', 'VALUE'])

# save the merged dataset to a new file if needed.
newdata_df.to_csv('newdata_df_dataset.csv', index=False)

#Merge by two or more different columns
merged_df = pd.merge(datapop_2016, datapop_2022, on = ['County', 'Administrative Counties 2019'])
merged_df
```

```python
datapop_2016 = pd.DataFrame({'Statistic Label': ['Population'],'Census Year': ['2016'], 'Sex': ['Males', 'Both Sexes', 'Females'],
                            'Marital Status': 'Single','Married', 'Separeted', 'Divorced', 'Widowed', 'Total'})
                             
datapop_2022 = pd.DataFrame({'Statistic Label': ['Population'],'Census Year': ['2022']})
  
# the default behaviour is join='outer'
# inner join
  
result = pd.concat([datapop_2016, datapop_2022], axis=1, join='inner')
display(result)
```
